```python
grocery = (gdf.groupby(['Member_number' , 'itemDescription'])['Date']
           .count().unstack().fillna(0))
```

```python
grocery
```

```python
grocery.shape
```

```python
grocery.columns
```

```python
def one_hot(x):
    if x <= 0:
       return 0
    else:
        return 1
```

```python
grocery = grocery.applymap(one_hot)
```

```python
grocery
```

```python
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
freq_items = apriori(grocery ,min_support= 0.05 , use_colnames= True)
```

```python
freq_items
```

```python
rules = association_rules(freq_items ,metric='lift' , min_threshold= 0.45)
```

```python
rules
```

```python
rules[rules['confidence'] >= 1 ]
```

```python
rules[rules['confidence'] >= 0.60 ]
```

```python
rules[rules['confidence'] >= 0.55 ]
```
